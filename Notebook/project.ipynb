```python
import pandas as pd
import numpy as np
```

```python

an_data = pd.read_csv("D:/project folder/secon_project/Luxury_Housing_Bangalore.csv")
an_data.head(10)

```

for inconsist format of ticket price

```python
an_data['Ticket_Price_Cr'] = (
    an_data['Ticket_Price_Cr']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace('Cr', '', regex=False)
    .str.strip()
)
```

```python
an_data.head(10)
```

```python
an_data.info()

```

for knull values

```python
# Fill numeric column Amenity_Score with median
an_data['Amenity_Score'] = an_data['Amenity_Score'].fillna(an_data['Amenity_Score'].median())

# Fill  column ticket_price_cr with medien
an_data['Ticket_Price_Cr'] = pd.to_numeric(an_data['Ticket_Price_Cr'], errors='coerce')
an_data['Ticket_Price_Cr'] = an_data['Ticket_Price_Cr'].fillna(an_data['Ticket_Price_Cr'].median())

#fill column unit size sqft with medien
an_data['Unit_Size_Sqft'] = an_data['Unit_Size_Sqft'].fillna(an_data['Unit_Size_Sqft'].median())

```

```python
an_data.head(10)
```

```python
an_data.info()
```

Normalize text fields

```python
# Convert text columns (Developer_Name, Micro_Market) to lowercase and strip spaces
text_cols = ['Developer_Name', 'Micro_Market']

for col in text_cols:
    an_data[col] = an_data[col].astype(str).str.strip().str.lower()

# Optional: remove special characters
an_data['Developer_Name'] = an_data['Developer_Name'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
an_data['Micro_Market'] = an_data['Micro_Market'].str.replace(r'[^a-zA-Z0-9 ]', '', regex=True)
```

```python
an_data.head(10)
```

derive column

```python
#  Price per Sqft
an_data['Price_per_Sqft'] = an_data['Ticket_Price_Cr'] / an_data['Unit_Size_Sqft']
#  Quarter Number from date column
an_data['Purchase_Quarter'] = pd.to_datetime(an_data['Purchase_Quarter'])
an_data['Quarter_Number'] = an_data['Purchase_Quarter'].dt.quarter
#  Booking Flag (1 = booked, 0 = not booked)

an_data['Booking_Flag'] = an_data['Possession_Status'].apply(
    lambda x: 1 if str(x).lower() == "ready to move" else 0
)
```

```python
an_data.head(10)
```

```python
an_data.to_csv("bangalore_luxury_housing.csv",index=False)
```

```python
import pandas as pd

cursor = mydb.cursor()

# Read CSV
df = pd.read_csv("bangalore_luxury_housing.csv")

# Replace NaN with None (for MySQL)
df = df.where(pd.notnull(df), None)

# ✅ Fix yes/no column
df['NRI_Buyer'] = (
    df['NRI_Buyer']
    .astype(str)
    .str.lower()
    .map({'yes': 1, 'no': 0})
    .fillna(0)
)

# ✅ Prepare SQL
columns = ", ".join(df.columns)
placeholders = ", ".join(["%s"] * len(df.columns))

sql = f"""
INSERT IGNORE INTO luxury_housing_bangalore ({columns})
VALUES ({placeholders})
"""

# ✅ Convert DataFrame to tuple list
data = [tuple(row) for _, row in df.iterrows()]

# ✅ Insert data
cursor.executemany(sql, data)

# ✅ Commit & close
mydb.commit()
cursor.close()
mydb.close()

print("✅ Data inserted successfully (duplicates ignored)")
```
